```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

El dataset del precio del Bitcoin con columnas como "Open", "High" y "Volume" se refiere a un conjunto de datos que registra información diaria u horaria del precio del Bitcoin en el mercado. Aquí está el significado de cada columna comúnmente utilizada en este tipo de dataset:

- "Open": Representa el precio de apertura del Bitcoin en un período de tiempo específico (día o hora). Indica el precio al comienzo de ese período.

- "High": Indica el precio más alto alcanzado por el Bitcoin durante el mismo período de tiempo. Es el precio máximo registrado.

- "Low": Representa el precio más bajo alcanzado por el Bitcoin durante el mismo período de tiempo. Es el precio mínimo registrado.

- "Close": Es el precio de cierre del Bitcoin en el período de tiempo dado. Indica el precio al finalizar ese período.

- "Volume": Hace referencia al volumen de transacciones de Bitcoin realizadas durante el período de tiempo especificado. Representa la cantidad total de Bitcoin que se compró y vendió.

Estas columnas permiten analizar y visualizar las fluctuaciones del precio, los niveles máximos y mínimos alcanzados, así como el volumen de transacciones realizadas.

Con esto podemos conocer con que frecuencia se mantuvo el precio del bitcoin a lo largo del día

```python
# Visualiza la distribución de una variable numérica con un histograma en relacion a la columna 'Open'
plt.hist(df['Open'])
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Distribución de la variable')
plt.show()
```

```python
# Visualiza la distribución de una variable numérica con un histograma en relacion a la columna 'High'
plt.hist(df['High'])
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Distribución de la variable')
plt.show()
```

```python


# Variables
fechas = df['Date']
valores = df['Open']

# Gráfico de tendencia con fechas ajustadas
plt.plot(fechas, valores)
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.title('Tendencia de la variable a lo largo del tiempo')

# Ajustar el intervalo de visualización de las fechas en el eje x
n = 25  # Número de fechas a mostrar en el eje x (ajústalo según tus preferencias)
plt.xticks(range(0, len(fechas), len(fechas)//n), fechas[::len(fechas)//n], rotation=45)

plt.show()

```
